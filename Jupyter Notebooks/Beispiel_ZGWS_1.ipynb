# Beispiel für den zentralen Grenzwertsatz

Wir simulieren den zentralen Grenzwertsatz für Summen. Dabei nehmen wir an, dass 5000 Arbeiter auf einer Baustelle eine uniforme Verteilung der Arbeitszeiten auf die ganzen Stunden hat (siehe Folion). Die Zufallsvariable $X$ ist dann der Fehler bis zur vollen Stunde und es gilt $X\sim\text{Unif}(-0.5,0.5)$.

Im folgenden sehen wir das Histogramm für die den Fehler nach einem Tag. 

```python
from scipy.stats import uniform
import numpy as np
import pandas as pd

n=1


x_1 = pd.Series(uniform.rvs(size=n*5000, loc=-0.5, scale =1)) 

x_1.plot(kind="hist", bins=np.arange(-2,2,0.25), edgecolor="black")
```

Für 2 Arbeitstage

```python
n=2

x_2 = pd.Series(uniform.rvs(size=n*5000, loc=-0.5, scale =1)) 

x_2 = pd.DataFrame(x_2.values.reshape(n,5000))
x_2_s = x_2.sum()

x_2_s.plot(kind="hist", bins=np.arange(-2,2,0.25), edgecolor="black")
```

Wir definieren nun eine _Funktion_ work_sum, die mit der Eingabe der Arbeitstage das entsprechende Histogramm plottet. Im Wesentlichen wird einfach der Index 1 und 2 durch den allgemeinen Index n ersetzt wird. Der Funktionsaufruf ist dann beispielsweise work_sum(16).

```python
def work_sum(n):
    x_n = pd.Series(uniform.rvs(size=n*5000, loc=-0.5, scale =1)) 
    x_n = pd.DataFrame(x_n.values.reshape(n,5000))
    x_n_s = x_n.sum()
    x_n_s.plot(kind="hist", bins=np.arange(-15,15,0.5), edgecolor="black")
    
work_sum(164)
```

Analog können wir eine Funktion work_mean() definieren.

```python
def work_mean(n):
    x_n = pd.Series(uniform.rvs(size=n*5000, loc=-0.5, scale =1)) 
    x_n = pd.DataFrame(x_n.values.reshape(n,5000))
    x_n_m = x_n.mean()
    x_n_m.plot(kind="hist", bins=np.arange(-0.5,0.5,.01), edgecolor="black")
    
work_mean(16)
```
