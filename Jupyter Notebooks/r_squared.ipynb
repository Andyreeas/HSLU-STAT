```python
import pandas as pd 
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import abline_plot
from scipy.stats import uniform, norm
import matplotlib.pyplot as plt
import numpy as np

x = uniform.rvs(size=20, loc=0, scale=10)
y = x 

df = pd.DataFrame({"x":x,"y":y})
df


Y = df["y"]
X = df["x"]
X = sm.add_constant(X)

fit = sm.OLS(Y,X).fit()

ax = df.plot(kind="scatter",x="x", y="y")
abline_plot(model_results=fit, ax = ax, color="orange")
plt.show()
```

```python
ax = df.plot(kind="scatter",x="x", y="y")
abline_plot(model_results=fit, ax = ax, color="orange")
for i in range(0,20):
    plt.plot([x[i],x[i]],[Y[i],np.mean(Y)], color="red")
    
plt.plot([0,10],[np.mean(y),np.mean(y)])
```

```python
df.corr()
```

```python
fit.rsquared
```

```python
Y.var()
```

```python
x = uniform.rvs(size=20, loc=0, scale=10)
y = x**2

df = pd.DataFrame({"x":x,"y":y})
df

ax = df.plot(kind="scatter",x="x", y="y")
Y = df["y"]
X = df["x"]
X = sm.add_constant(X)
fit = sm.OLS(Y,X).fit()
abline_plot(model_results=fit, ax = ax, color="orange")
weights = np.polyfit(x, y, 2)
model = np.poly1d(weights)
fit = sm.OLS(Y,model(X)).fit()

x_g = np.linspace(0,10)
y_g = x_g**2



plt.plot(x_g,y_g, linewidth=.5)
plt.show()
```

```python
df.corr()
```

```python
fit.rsquared
```

```python
Y.var()
```
