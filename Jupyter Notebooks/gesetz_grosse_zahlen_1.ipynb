Wir w체rfeln $n=10$ und $n=40$ mal und z채hlen die Augensumme $S_n$ und den Durchschnitt $\overline{X}_n$. Dies machen wir 1000 mal und plotten das Histogramm. 

F체r die Summe erhalten wir folgende Histogramme: 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

trials = 1000
n_1 = 10
n_2 = 40


x_1 = pd.Series(np.random.randint(1, 7, trials*n_1))


x_2 = pd.Series(np.random.randint(1, 7, trials*n_2))

mat_1 = pd.DataFrame(x_1.values.reshape(n_1, trials))
mat_2 = pd.DataFrame(x_2.values.reshape(n_2, trials))

x_1_s = mat_1.sum()
x_2_s = mat_2.sum()
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.xlim(0,100)
x_1_s.plot(kind="hist", edgecolor="black")

plt.subplot(122)
plt.xlim(100,200)
x_2_s.plot(kind="hist", edgecolor="black")
```

F체r den Durchschnitt sehen die Histogramme wie folgt aus:

```python
x_1_m = mat_1.mean()
x_2_m = mat_2.mean()

plt.figure(figsize=(15, 4))

plt.subplot(121)
plt.xlim(1,5)
x_1_m.plot(kind="hist", edgecolor="black")

plt.subplot(122)
plt.xlim(1,6)
x_2_m.plot(kind="hist", edgecolor="black")
```
